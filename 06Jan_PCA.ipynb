```python
import pandas as pd
df=pd.read_csv('wine.data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
x=df.drop(('Class'),axis=1)
y=df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x=pd.DataFrame(data=x,columns=df.columns[1:])
```

```python
x.head()
```

```python
y.head()
```

#PCA

```python
from sklearn.decomposition import PCA
pca=PCA()
pca_array=pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_
```

```python
pca_array=pd.DataFrame(data=pca_array,columns=df.columns[1:])
pca_array.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=[i+1 for i in range(len(pca.explained_variance_ratio_))],y=pca.explained_variance_ratio_)
plt.grid()
plt.show()
```
