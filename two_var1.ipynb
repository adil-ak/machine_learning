```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_excel("two_variable.xlsx")
df
```

```python
model=linear_model.LinearRegression()
model.fit(df[["area"]],df.price)
```

```python
model.predict([[5000]])
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("Price")
plt.scatter(df.area,df.price,color="red",marker="+")
plt.plot(df.area,model.predict(df[["area"]]),color='blue')
```

```python
d=pd.read_excel('predict_twovar.xlsx')
d
```

```python
p=model.predict(d)
p
```

```python
d['price']=p
d
```

```python
d.to_excel('predicted_twovar.xlsx')
```
