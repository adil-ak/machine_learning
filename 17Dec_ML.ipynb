Confusion Matrix

**Accuracy :**
The proportion of correct predictions (both positive and negative) out of the total predictions.

Equation : Accuracy = (TP+TN)/(FP+FN+TP+TN)


**Precision :**
The Proportion of true positive predictions out of all positive predictions.

Equation : Precision = TP/(TP+FP)


**Recall :**
The proportion of actual positives that are correctly predicted.

Equation : Recall=TP/(TP+FN)

**F1 Score:**
The harmonic mean of precision and recall, balancing the two matrices.

Equation : F1 Score= 2*((Precision*Recall)/(Precision+Recall))



   High Precision means fewer false positives,
   High Recall means fewer false negatives
